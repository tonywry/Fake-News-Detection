```python
from googlesearch import search
import urllib.request
from bs4 import BeautifulSoup
import re

def google_search(query, top_n):
    title_list = []
    for url in search(query, stop=top_n):
        if url[-3:] not in ("pdf",".pl") and "term-graph.html" not in url:
            print("Scraping:", url)
            try:
                page = urllib.request.urlopen(url, timeout=15)
                content = str(page.read())
                soup = BeautifulSoup(content)
                title_string = re.sub('[^a-z A-Z1-9]+', '', soup.title.string)
                title_list.append(title_string)
                print(title_string)
            except Exception as e:
                print("error:", e)
                continue
    return title_list
```

### Scrap titles and save in csv

```python
import time
for index in [6623,6626,7331,7393,7973,8020,8319,8671]:
    try:
        print("Scraping", index)
        query = fake_news["Statement"][index]
        search_result = google_search(query, 30)
        searched_df =  pd.DataFrame({"searched":search_result})
        searched_df.to_csv("Google_Scraped/title_scraped_"+str(index)+".csv")
    except:
        time.sleep(60*5)
```
